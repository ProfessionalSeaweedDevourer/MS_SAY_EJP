## Pivot
Pivot은 개발자의 기술 스택과 활동 데이터를 실시간으로 수집하여 시각화하는 인터랙티브 대시보드입니다. 기술 간의 상관관계와 숙련도의 변화를 동적으로 추적합니다.

#### 1. 핵심 기능 (Core Features)
* Interactive Tech-Radar: D3.js를 활용하여 보유 기술의 숙련도와 활용 빈도를 좌표 평면 위에 시각화합니다.
* Live GitHub Activity: GitHub API를 통해 실시간 커밋 내역 및 언어 사용 비중을 분석하여 반영합니다.
* Skill Connectivity: 연관된 기술(예: React - TypeScript) 간의 연결망을 그래프 형태로 출력합니다.
* Performance Metrics: Lighthouse 지표 및 프로젝트 성능 데이터를 대시보드 형태로 제공합니다.

#### 2. 프로젝트 구조 및 주요 스크립트 (Architecture)
```
src/
├── api/            # 외부 API(GitHub 등) 통신 로직
├── components/     # UI 공통 컴포넌트 및 시각화 모듈
├── hooks/          # 데이터 페칭 및 상태 관리를 위한 Custom Hooks
├── utils/          # 데이터 가공 및 수식 처리 함수
└── App.jsx         # 애플리케이션 엔트리 포인트
```
```
| 스크립트 | 역할 설명 |
|fetchStats.js | GitHub API를 호출하여 사용자 활동 데이터를 정제합니다. |
| drawRadar.js | D3.js를 사용해 기술 숙련도 레이더 차트를 렌더링합니다. |
| useTransform.js | 로우 데이터를 시각화 라이브러리에 적합한 포맷으로 변환합니다. |

#### 3. 기술 스택
