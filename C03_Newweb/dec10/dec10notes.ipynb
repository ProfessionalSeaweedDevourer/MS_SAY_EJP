## 오늘의 준비물

#### 애드온 관리자 yarn 설치
```npm install yarn -g```

#### React 대비 VSC 확장 설치
* ESLint: JS계 자동완성
* Prettier: JS계 코드 정렬
* ES7 + React : React계 자동완성

## 주요 파일 내용

* ```public/index.html```: ```<div id="root"></div>``` Root 영역 존재
* ```src/index.js```: 기본 설정 & Root에 ```App.js``` 띄우기
* ```src/App.js```: 실제 작업 영역
