## 준비물

#### 애드온 관리자 yarn 설치
```npm install yarn -g```

#### React 대비 VSC 확장 설치
* ESLint: JS계 자동완성
* Prettier: JS계 코드 정렬
* ES7 + React : React계 자동완성

## 주요 파일 내용

* ```public/index.html```: ```<div id="root"></div>``` Root 영역 존재
* ```src/index.js```: 기본 설정 & Root에 ```App.js``` 띄우기
* ```src/App.js```: 실제 작업 영역
* ```src/App.css```: 실제 작업 영역의 디자인

## React 프로젝트 생성
```yarn create vite```

프로젝트명 > React > JS > no > yes

#### 프로젝트 실행 / 빌드
```cd 프로젝트명```

```yarn start```
* package.json 내부 script.start에 있는 명령어를 실행.

##### **테스트 빌드에서 80번 포트를 쓰는 이유**
HTTP 기본 포트가 80번이라, 따로 포트 지정 안 하면 바로 연결.

##### **8080을 대신 쓰는 이유**
Linux에서는 1024번 이하의 포트는 root 권한 있어야 바인딩 가능해서 불편함

OracleDB도 8080 씀(아오)

## 클래식 웹 vs 신형 웹

클래식 웹
 * 메인 서버 하나에 부하가 너무 많이 걸림
 * 분업화의 어려움: 데이터를 여러 군데에서 가져오기에는 신형 웹이 적합
