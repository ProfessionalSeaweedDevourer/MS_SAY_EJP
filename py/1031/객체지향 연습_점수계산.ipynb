### 객체지향 연습
간단한 학생 성적 데이터(이름, 점수)를 받아 평균을 계산하고, 최고/최저 점수를 찾는 $\texttt{ScoreAnalyzer}$ 클래스를 설계해 보세요.

##### 1단계) 원시 알고리즘

임의의 튜플로 입력 데이터를 받음 - 최고/최저점 결과 출력을 위해, 각 점수는 이름과 엮여 있어야 함.

1) 이 입력 과정은 특정 입력이 들어올 때까지 반복해야 함. try-except로 구현
2) 입력값 검증 수행. 이름에 텍스트가 아닌 것이 들어오는 것은 허용해야 할지? 일단 점수에는 무조건 정수 내지 float만 들어가야 함.
3) 정상적이지 않은 입력이 발생하면 다시 입력으로 보냄. continue 활용?

>> 튜플들을 최종적으로 담을 메인 컨테이너는 **리스트(List)**가 적절합니다. $\texttt{List}$는 새로운 성적 튜플을 계속 추가($\texttt{append}$)할 수 있기 때문입니다.
>> '이름에 텍스트가 아닌 것' 허용 여부는 결정이 필요합니다. 일반적으로는 문자열만 허용하는 것이 논리적입니다. $\texttt{continue}$를 사용하여 잘못된 입력 시 루프의 처음으로 돌아가는 것은 올바른 제어 흐름입니다.

입력이 마무리되면, 대상 튜플들로부터 평균 점수를 계산.

1) 반복문을 활용해, 전체 튜플 개수 횟수만큼 각 튜플에서 점수 입력값 부분을 받고, 이를 특정 변수에 더해 나감.
2) 반복이 끝난 후, 점수 총합이 되었을 이 변수 값을 튜플 개수로 나눔. 입력 검증은 1.2에서 수행했으므로 여기서는 문제가 없을 것.

>> Python에서는 sum() 내장 함수와 **컴프리헨션(Comprehension)**을 사용하여 반복문을 명시적으로 사용하지 않고도 총합을 매우 간결하게 계산할 수 있습니다. (예: $\texttt{sum(score for name, score in data\_list)}$)
>> 나누기 전에 튜플 개수가 0인지(즉, 입력된 데이터가 없는지) 확인하여 **$\texttt{ZeroDivisionError}$**를 방지하는 예외 처리를 추가해야 합니다.

최고점과 최저점 구하기.
1) 또 반복문을 사용해 전체 튜플들에 다시 하나씩 접근할 수도 있지만...이 기능을 평균 점수 구하는 반복 과정과 통합하는 게 처리 성능이 더 낫지 않을까? (보류)
>> 좋은 고민. 평균 계산과 최고/최저점 탐색을 하나의 반복문에서 처리하는 것이 처리 성능(반복 횟수) 측면에서 가장 효율적입니다.
>> 강력히 권장. 하나의 반복문에서 점수 누적, 최대값 갱신, 최소값 갱신을 동시에 처리하도록 로직을 통합해야 합니다.

2) 반복문을 활용해서, 모든 튜플의 점수 부분을 모은 새로운 하나의 튜플을 생성...그런데 이런 걸 만들 거면 이걸 그대로 평균 계산에도 쓸 수 있지 않나?

>> 불필요. 점수만 모은 새로운 리스트/튜플을 생성하는 것은 메모리 낭비이며, 점수에 해당하는 이름 정보가 유실될 위험이 있어 최고/최저점 학생을 찾을 수 없습니다.

3) 정렬이나 탐색을 사용해서 최대, 최소점을 찾는 것 자체는 이미 존재하는 함수로 간단하게 가능.

>> 유효함. $\texttt{max()}$, $\texttt{min()}$ 함수를 사용하는 것은 빠르고 간결합니다.
>> 이름과 엮여 있는 튜플 리스트에서도 key 인자를 사용하여 점수($\texttt{tuple[1]}$)를 기준으로 최대/최소 튜플을 찾을 수 있습니다. (예: $\texttt{max(data\_list, key=lambda x: x[1])}$)

4) 이 찾은 값이 '몇 번째 원소인지'를 함께 표시해 줘야 이게 누구의 성적인지 알 수 있으므로, 정렬보다는 이쪽으로 해결하는 것이 논리적으로 맞아 보임.

>> 핵심. 학생 정보(이름)를 함께 출력하려면 점수뿐만 아니라 튜플 자체를 찾아야 한다는 논리는 완벽하게 맞습니다.
>> max(data_list, key=...) 방법을 사용하면 점수를 기준으로 최대/최소 튜플을 찾을 수 있으므로, 해당 튜플($\texttt{('이름', 점수)}$)에서 이름 정보를 쉽게 추출할 수 있습니다.

5) 다만 정렬이 처리 속도는 훨씬 빠를 것 같은데...이 부분은 아직 잘 모르겠음.

>> 정렬($\mathcal{O}(N \log N)$) vs. 순차 탐색($\mathcal{O}(N)$). 최고/최저점만 구하는 목적이라면, 정렬보다 순차 탐색이 훨씬 빠릅니다.
>> 순차 탐색을 사용하세요. Python의 내장 $\texttt{max/min}$ 함수($\mathcal{O}(N)$) 또는 단일 통합 반복문($\mathcal{O}(N)$)을 사용하는 것이 정렬($\mathcal{O}(N \log N)$)보다 성능상 이득입니다.

평균, 최고점과 해당 학생, 최저점과 해당 학생을 출력.

제시된 논리를 바탕으로 $\texttt{ScoreAnalyzer}$의 1단계 알고리즘은 다음과 같이 정리될 수 있습니다.

데이터 저장: 성적 데이터를 담을 빈 리스트($\texttt{scores\_data}$)를 준비합니다.

입력 및 검증:$\texttt{while True}$ 루프를 돌며 이름과 점수를 입력받습니다.
    이름이 $\texttt{STOP}$이면 $\texttt{break}$합니다.
    점수를 $\texttt{try-except}$로 $\texttt{float}$으로 변환하여 검증하고, 실패 시 $\texttt{continue}$합니다.
    성공 시, $\texttt{(이름, 점수)}$ 튜플을 $\texttt{scores\_data}$에 추가($\texttt{append}$)합니다.

결과 계산:데이터 리스트가 비어있지 않은지 확인합니다.
$\texttt{max()}$ / $\texttt{min()}$ 함수를 $\texttt{key}$와 함께 사용하여 최고점 튜플과 최저점 튜플을 직접 구합니다.
    (이름 정보 유지)$\texttt{sum()}$ 함수와 $\texttt{len()}$을 사용하여 평균을 계산합니다.

출력: 계산된 평균, 최고점/학생, 최저점/학생을 출력합니다.

### 2단계) 메서드 설계
##### 단일 책임 원칙 - 메서드당 역할은 하나씩, 확실하게.

데이터 저장: 성적 데이터를 담을 빈 리스트를 생성하는 메서드.

>> 불필요. 데이터 컨테이너(리스트)는 클래스 인스턴스가 생성될 때, 즉 **생성자 $(\texttt{\_\_init\_\_})$**에서 초기화하는 것이 표준입니다. 객체가 생성될 때 데이터를 담을 준비를 하는 것이 자연스럽습니다.
>> 생성자 $(\texttt{\_\_init\_\_})$ 내에서 $\texttt{self._scores_data = []}$와 같이 초기화합니다. (데이터 은닉을 위해 $\texttt{\_}$ 사용)

입력 및 검증: 생성된 리스트에 루프를 통해 입력받은 이름과 점수 데이터를 저장. 이름에서 STOP이 들어오면 break, 점수도 검증하고 저장.

>> 적절함. 루프를 돌며 사용자 입력을 받고, 점수를 검증하며 데이터를 내부 리스트에 저장하는 것은 이 메서드의 유일한 책임으로 적절합니다.
>> 데이터 입력/저장과 검증 로직을 한 메서드에 두는 것은 효율적입니다. 이 메서드는 사용자 입력 처리가 끝났음을 알리는 상태 변수를 갱신하도록 설계할 수 있습니다. (예: $\texttt{self.\_is\_data\_collected = True}$)

결과 계산: 데이터가 비어있지 않은지 확인 - 이것은 빈 칸 데이터 입력이나 추후 데이터 수정의 여지 때문인지? 입력 검증 메서드를 거치는 설계 구조상 필요성이 잘 느껴지지 않음. max/min 함수로 최고/최저점 튜플을 구하고, sum과 len 함수로 평균 계산

>> 필요함. 입력 메서드를 거쳤더라도, 사용자가 $\texttt{STOP}$을 입력하여 데이터가 하나도 저장되지 않은 채 메서드가 호출될 수 있습니다. $\texttt{len(\dots)}$이 0일 때 평균 계산 시 $\texttt{ZeroDivisionError}$가, $\texttt{max()}$ 호출 시 $\texttt{ValueError}$가 발생합니다. 방어 코드로 반드시 필요합니다.
>> 분리 필요. 현재 설계는 '평균 계산'과 '최고/최저점 찾기'라는 두 가지 독립적인 분석 결과를 한 메서드에서 모두 계산하고 있습니다. SRP에 따라 더 세분화해야 합니다.
>> 세분화:1. $\texttt{get\_average(self)}$: 평균 값만 계산하여 반환2. $\texttt{get\_highest\_scorer(self)}$: 최고점 튜플만 반환3. $\texttt{get\_lowest\_scorer(self)}$: 최저점 튜플만 반환

출력: 최고, 최저점과 그 대상 학생, 그리고 평균 점수를 출력.

>> 적절함. 앞서 분리된 계산 메서드들을 호출하여 데이터를 받고, 이를 포맷하여 사용자에게 보여주는 표현(Presentation) 책임만 갖습니다.
>> $\texttt{display\_analysis(self)}$ 또는 $\texttt{print_results(self)}$처럼 동사로 시작하는 것이 메서드 이름 관례상 더 좋습니다.

메인 루프: 위 메서드를 순서대로 호출.

>> 적절함. $\texttt{if \_\_name\_\_ == "\_\_main\_\_":}$ 블록에서 클래스 인스턴스를 생성하고, 데이터 수집 -> 계산 및 분석 -> 출력의 순서로 메서드를 호출하는 것은 자연스러운 흐름입니다.

```python
# 완성 코드 샘플

import sys

class ScoreAnalyzer:
    
    def __init__(self):
        # 성적 데이터 (이름, 점수) 튜플을 저장할 리스트
        self._scores_data = []
        # 데이터 수집 완료 상태를 추적
        self._is_data_collected = False 

    def collect_scores(self):
        print("성적 입력 모드: 이름과 점수를 입력하세요. 이름에 'STOP' 입력 시 종료됩니다.")
        
        while True:
            try:
                name = input("이름: ").strip()
                
                if name.upper() == "STOP":
                    break
                
                score_input = input("점수: ").strip()
                
                # 점수 검증 및 변환
                score = float(score_input)
                
                if score < 0:
                    print("점수는 음수가 될 수 없습니다. 다시 입력해 주세요.")
                    continue
                    
                # 데이터 저장
                self._scores_data.append((name, score))
                
            except ValueError:
                print("유효하지 않은 점수 입력입니다. 점수는 숫자여야 합니다.")
                continue
            except EOFError:
                # 터미널에서 Ctrl+D (Linux/macOS) 또는 Ctrl+Z (Windows) 입력 시 종료
                print("\n입력 스트림 종료 감지. 수집을 마칩니다.")
                break
            except Exception as e:
                print(f"예상치 못한 오류 발생: {e}")
                break
        
        self._is_data_collected = True
        if not self._scores_data:
            print("저장된 성적 데이터가 없습니다.")

    # -----------------------------------------------------
    # 분석 메서드: 데이터가 없을 경우 처리
    # -----------------------------------------------------

    def _check_data(self):
        """데이터가 비어있는지 확인하고, 비어있으면 예외를 발생시킵니다."""
        if not self._is_data_collected:
            print("경고: 먼저 collect_scores()를 실행해야 합니다.", file=sys.stderr)
            return False
            
        if not self._scores_data:
            print("경고: 분석할 성적 데이터가 없습니다.", file=sys.stderr)
            return False
            
        return True

    def get_average(self):
        if not self._check_data():
            return 0.0
            
        # 점수만 추출하여 합계를 계산
        total_score = sum(score for name, score in self._scores_data)
        count = len(self._scores_data)
        
        # ZeroDivisionError 방지는 _check_data에서 처리되지만, 방어적으로 한 번 더 확인
        return total_score / count if count > 0 else 0.0

    def get_highest_scorer(self):
        if not self._check_data():
            return None, None
        
        # lambda 함수를 사용하여 튜플의 두 번째 요소(점수)를 기준으로 최대 튜플을 찾음
        highest_tuple = max(self._scores_data, key=lambda x: x[1])
        return highest_tuple
    
    def get_lowest_scorer(self):
        if not self._check_data():
            return None, None
            
        # lambda 함수를 사용하여 튜플의 두 번째 요소(점수)를 기준으로 최소 튜플을 찾음
        lowest_tuple = min(self._scores_data, key=lambda x: x[1])
        return lowest_tuple

    # -----------------------------------------------------
    # 표현 메서드
    # -----------------------------------------------------

    def display_analysis(self):
        if not self._check_data():
            return

        print("\n--- 성적 분석 결과 ---")
        
        try:
            average = self.get_average()
            highest_name, highest_score = self.get_highest_scorer()
            lowest_name, lowest_score = self.get_lowest_scorer()
            
            print(f"총 학생 수: {len(self._scores_data)}명")
            print(f"평균 점수: {average:.2f}점")
            print("-" * 20)
            print(f"최고 득점자: {highest_name} ({highest_score:.2f}점)")
            print(f"최저 득점자: {lowest_name} ({lowest_score:.2f}점)")

        except Exception as e:
            print(f"분석 중 오류 발생: {e}")
            
# =========================================================
# 메인 실행 루프 (Main Loop)
# =========================================================
if __name__ == "__main__":
    
    analyzer = ScoreAnalyzer()
    
    # 1. 데이터 수집
    analyzer.collect_scores()
    
    # 2. 결과 분석 및 출력
    analyzer.display_analysis()
```
