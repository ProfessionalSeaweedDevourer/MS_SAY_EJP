### **네이버 뉴스 API 받아오기**
>[Naver Devs](https://developers.naver.com)

```python
from http.client import HTTPSConnection
from urllib.parse import quote

q1 = "경제"
q1 = quote(q1)
print(q1)

k = {"X-Naver-Client_Id" : "Jw3irlagRSw_EoSLFaC6",
     "X-Naver-Client-Secret" : "zDIWqwbkyK"}

hc = HTTPSConnection("openapi.naver.com") # '대상 폴더' 직전까지의 도메인, 또는 ip, 포트 등
hc.request("GET", "/v1/search/news.xml?query="+q1, headers = k)

res = hc.getresponse()
resBody = res.read()
print(resBody.decode())

hc.close()
```
