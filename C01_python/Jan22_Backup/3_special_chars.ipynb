```python
from sys import maxunicode
from unicodedata import category


special = []
for i in range(maxunicode):
    if category(chr(i)).startswith("P"):
        special.append(chr(i))
```

```python
f = open("./Users/student02_04/Jan22_1_Text/ThreeKingdoms_01.txt", "r", encoding="euc-kr")

# 단어 개수 세기
wc ={}
for line in f.readlines():

    for s in special:
        line = line.replace(s, "")
        # 특수문자를 먼저 지운다.   

    line = line.replace("\n", "").split(" ")
    for word in line:
        if word in wc:
            wc[word] += 1
        else:
            wc[word] = 1
f.close()

# for w, c in wc.items():
#     print(w, c)

# 결과를 정렬하는 것도 필요할 것. '가장 자주 등장하는 단어 순'으로 정렬

wc_sorted = sorted(wc.items(), key=lambda x: x[1], reverse=True)
for w, c in wc_sorted:
    print(w, c)
```
