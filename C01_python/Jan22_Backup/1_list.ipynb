```python
a = ["ㄱ", "ㄴ", "ㄷ"]
b = ["ㄹ", "ㄴ", "ㅁ", "ㄷ"]

```

```python
# 1) b의 원소 중 a에도 있는 원소만을 선택해 출력
# python 기초 기능만을 사용

for element in b: # b의 각 원소에 대해...
    if element in a: # '만약' 그 원소가 a에도 있다면, 그 경우에만...
        print(element) # 그 원소를 출력
```

```python
# 2) b의 모든 원소를 갖는 새로운 리스트 생성
c = []
for element in b:
    c.append(element)
print(c)

# 또는
d = [element for element in b]
print(d)
```

```python
# 3) b 원소 중 a에도 있는 것만을 원소로 갖는 새로운 리스트 생성

e =[]
for element in b:
    if element in a: # 'a에도 있는 것에 한해서'
        e.append(element) # e의 원소로 추가

print(e)

# 또는
f = [element for element in b if element in a] # 파끼야아아악
print(f)
```
