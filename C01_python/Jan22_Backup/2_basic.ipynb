```python
f = open("./Users/student02_04/Jan22_1_Text/ThreeKingdoms_01.txt", "r", encoding="euc-kr")

# 단어 개수 세기
wc ={}
for line in f.readlines():
    line = line.replace("\n", "").split(" ")
    # 그러나 이런 식의 분리는 ","가 붙은 단어를 다른 단어로 취급하게 되는 문제가 있다.
    # => 문장부호, 특수문자 제거가 필요할까?
    for word in line:
        if word in wc:
            wc[word] += 1
        else:
            wc[word] = 1
f.close()

for w, c in wc.items():
    print(w, c)
```
