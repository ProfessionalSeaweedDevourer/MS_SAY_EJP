```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import os
from konlpy.tag import Okt

# 1. 환경 설정 및 객체 초기화
o = Okt()
target_dir = "./Users/student02_04/Jan23_1_Text"
file_path = os.path.join(target_dir, "naver_news_titles.txt")
font_path = os.path.join(target_dir, "MALGUN.TTF")

# 2. 불용어(Stopwords) 정의
# 의미가 미미한 동사 및 일반 명사 제외
stopwords = ['하다', '있다', '되다', '보다', '나타나다', '오다', '가다', '나다', '들다', '것', '수', '이', '때', '뉴스', '네이버']

wc = {}

# 3. 데이터 로드 및 분석
if os.path.exists(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f.readlines():
            line = line.strip()
            if not line or line.startswith("==="):
                continue
            
            line = o.normalize(line)
            # 명사(Noun)와 동사(Verb) 모두 추출
            for w, p in o.pos(line, stem=True):
                if p in ["Noun", "Verb"]:
                    # 불용어 체크 및 단어 길이 필터링 (한 글자 명사 제외 등)
                    if w not in stopwords and len(w) > 1:
                        wc[w] = wc.get(w, 0) + 1

    # 4. 시각화 데이터 준비 (상위 20개)
    sorted_wc = sorted(wc.items(), key=lambda x: x[1], reverse=True)[:20]
    words = [item[0] for item in sorted_wc]
    counts = [item[1] for item in sorted_wc]

    # 5. 시각화 실행
    if words:
        font_prop = fm.FontProperties(fname=font_path)
        x_pos = np.arange(len(words))
        
        plt.figure(figsize=(14, 7))
        bars = plt.bar(x_pos, counts, color='teal', alpha=0.7, edgecolor='black')
        
        # 축 및 제목 설정
        plt.xticks(x_pos, words, rotation=45, fontproperties=font_prop, fontsize=11)
        plt.title('News Keywords: Nouns & Verbs Top 20', fontproperties=font_prop, fontsize=16)
        plt.ylabel('Frequency', fontproperties=font_prop)
        
        # 막대 위에 빈도수 표시
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.2, int(yval), ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.savefig(os.path.join(target_dir, "news_keywords_analysis.png"))
        plt.show()
    else:
        print("분석된 단어가 없습니다.")
else:
    print(f"파일을 찾을 수 없습니다: {file_path}")
```
