```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()
df["fight"] = [80, 95, 10, 90, 5]
df["yok"] = [20, 5, 90, 10, 95]

# print(df)

sns.scatterplot(df, x="fight", y="yok")
plt.show()
```

```python
# kMeans: K개의 '평균들' => 군집화

# 1) 가령 k=2일 때, 무작위 2개 점을 초기 중심점으로 선택
# 2) 각 데이터포인트를 유클리드 거리 기준으로 1)에서 생성한 중심점 중 가까운 것에 할당
# 3) 2)에서 얻은 데이터의 평균을 계산해 중심점을 적절한 곳으로 업데이트
# 4) 3)을 반복. 중심점의 위치가 반복해도 그대로거나 최대 반복 횟수에 도달하면 종료

from sklearn.cluster import KMeans

data = df[["fight", "yok"]].to_numpy()

km = KMeans(2)
# print(km.fit_predict(data))
df["group"] = km.fit_predict(data)

sns.scatterplot(df, x="fight", y="yok", palette="rainbow", hue="group")
plt.show()
```
