```python
# 이미지 데이터에 대한 학습

import tensorflow as tf
import numpy as np
import os
import cv2 
```

```python
def getImg(folder, w, h):
    datas = []
    for f in os.listdir(folder):
        data = cv2.imread(folder + f, cv2.IMREAD_GRAYSCALE)
        data = cv2.resize(data, (w, h))
        datas.append(data)
    return np.array(datas)
```
