```python
import torch

xData = torch.FloatTensor([[1], [2], [3]])
yData = torch.FloatTensor([[11], [21], [31]])
```

```python
W = torch.zeros(1, requires_grad=True) # 1차원의 GD 대상
b = torch.zeros(1, requires_grad=True)

o = torch.optim.SGD([W, b], lr = 0.01)
```

```python
for epoch in range(1000):
    f = xData * W + b
    l = torch.mean((f-yData) ** 2) # MSE

    o.zero_grad() # zero grad란? => 
    l.backward() # 역전파
    o.step() # 

    print(f" y = {W.item()}x + {b.item()}")
```
