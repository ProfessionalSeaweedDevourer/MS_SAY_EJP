```python
import torch

W = torch.tensor(5.0, requires_grad=True)
o = torch.optim.SGD([W], lr=0.01)
```

```python
for epoch in range(10):
    f = W * 2

    f.backward() # 역전파 => 지금은 단일 계층이므로 그냥 gradient
    print(W.grad.item())
    # 결과는 왜 2가 아니라 2n이 될까?
    # => backward()는 기존.grad에 연산 결과를 '누적'한다.
    # zerograd를 하지 않으면 이렇게 된다 => 메모리 부족 상황에서 데이터를 '나누어' 학습하기 위해 만든 기능
```
