```python
# 선형회귀

import tensorflow as tf

tf.compat.v1.disable_v2_behavior() # 여기까지가 템플런임
```

```python
# 1) 학습 데이터 확보 & 전처리
xData = [1, 2, 3]
yData = [13, 23, 33] # 편의상 최대한 간단하게
```

```python
# 2) 인공신경망의 구성
# 변수: variable (찾아내야 하는 값) / placeholder (학습데이터 저장)

# 초기값 설정: 0, 1차원, float
# ...의 방법론도 있지만 아무튼 지금은 단순하게
a = tf.Variable(tf.zeros([1], dtype=tf.float64))

# 초기값 설정: 정규분포에 따른 -1 ~ 2 사이의 무작위 1차원값
b = tf.Variable(tf.compat.v1.random_uniform([1], -1, -2, dtype=tf.float64))

x = tf.compat.v1.placeholder(tf.float64)
y = tf.compat.v1.placeholder(tf.float64)

f = x * a + b # a = 0, 무작위 b로부터 시작하는 초기 '식'
# a, b 값을 변화시키며 반복을 진행: 실제 y값에 접근할 때까지.

loss = tf.reduce_mean(tf.square(y-f)) # mse를 오차함수로 채택, 차원수 조정

o = tf.compat.v1.train.GradientDescentOptimizer(learning_rate = 0.1) # 최적 a,b를 찾아내기 위한 optimizer; GD 활용

# 목표 설정: loss 최소화
goal = o.minimize(loss)

```

```python
# 3) 학습 진행

s = tf.compat.v1.Session()
s.run(tf.compat.v1.global_variables_initializer()) # variable 무작위로 초기화

for i in range(10000): # 10,000회 반복 시행
    s.run(goal, feed_dict={x:xData, y:yData}) # 각 placeholder에 xData와 yData 투입
    # 매 반복 과정을 볼 '필요'는 없으나, 학습 목적이므로 다음을 추가
    # print(s.run(loss, feed_dict={x:xData, y:yData}))
    print(f"y = {s.run(a)}x + {s.run(b)}")
```

```python
# 4) 완성된 신경망을 활용, 미지의 데이터에 대한 결과 예측

newX = float(input("x : "))
newX = [newX]
result = s.run(f, feed_dict={x:newX})[0]
print(result)
```
