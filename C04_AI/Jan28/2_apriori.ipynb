```python
# %pip install apyori

trainData = [
    ["치킨", "치킨", "맥주", "소주", ],
    ["치킨", "맥주", "소주", ],
    ["치킨", "피자", "맥주", "소주", ],
    ["치킨", "막걸리", ],
    ]
```

```python
import re
from apyori import apriori

result = apriori(trainData, min_support=0.4, min_confidence=0.5)
result = list(result)

# * collection에 대해...
# array는 '수정'할 수 없다.
# list는 append 적용이 가능하다.
# frozenset: 수정이 불가능한 set

interpret_lift = lambda l: "양의 상관관계 (연관성 높음)" if l > 1 else ("독립적 관계" if l == 1 else "음의 상관관계 (연관성 낮음)")

for r in result:
    for r2 in r.ordered_statistics:
        print(r2.items_base, "구매자의")
        print(r2.items_add, "구매 확률: ")
        print(r2.confidence)
        print(f" 향상도(Lift): {r2.lift:.2f} -> {interpret_lift(r2.lift)}")
        # lift: apyori 라이브러리의 기능.
        # A와 B가 독립적일 때보다 얼마나 더 자주 '같이 발생'하는지 => Lift > 1이면 양의 상관관계가 있다고 판단.
        # '둘 다 살 확률' / ('A 살 확률' * 'B 살 확률')
        print("-----")
```
