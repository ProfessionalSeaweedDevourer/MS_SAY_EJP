```python
import tensorflow as tf

tf.compat.v1.disable_v2_behavior() # 2버전대 기능 강제 해제 테스트
```

```python
# v1 구동 테스트

node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0) 
node3 = tf.add(node1, node2)

# 세션을 열고 그래프를 실행
sess = tf.compat.v1.Session()

print("node3 실행 결과:", sess.run(node3)) # 결과값 7.0이 나오면 정상 작동 중입니다.

sess.close()
```

```python
# TensorFlow의 구동
# 1) 인공신경망 구성
# 2) 학습 데이터를 활용한 최적화
# 3) 완성된 신경망을 통해 연산


# 1) 인공신경망의 구성
a = tf.constant(10) # a,b,c는 하나의 'node'.
# 실제로 값 10이 '삽입'되지는 않음. 
print(a) # Tensor("Const_4:0", shape=(), dtype=int32)
b = tf.constant(20)
c = tf.add(a, b)
# 실제로 a+b가 '연산'되지는 않음.
d = tf.constant("N")

print(a,c,d) # 이는 각 '노드 정보'를 출력함.

# 2) 학습 데이터를 활용한 최적화
s = tf.compat.v1.Session() # 실행 '세션'의 생성.

aResult = s.run(a)
print(aResult) # a 노드의 '실행'으로 인해 결과 10을 얻음.

dResult = s.run(d) # d 노드의 실행.
print(dResult) # b'N': 바이트 문자 N임을 의미

cResult = s.run(c) # c 노드의 실행: c의 '구성'에 필요한 a, b도 함께 실행됨.
print(cResult)
```
