```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. 학습 데이터 준비 (문장과 라벨: 1은 칭찬, 0은 일반)
train_sentences = [
    "Your work is excellent.", 
    "This is a great job.",
    "It is raining today.",
    "I am going to the store."
]
labels = [1, 1, 0, 0]

# 2. 텍스트를 수치 데이터(벡터)로 변환
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_sentences)

# 3. 나이브 베이즈 모델 학습
model = MultinomialNB()
model.fit(X_train, labels)

# 4. 생성한 문장으로 테스트
test_sentence_1 = ["The presentation was outstanding."]
test_sentence_2 = ["The presentation you delivered today was absolutely outstanding and very professional."]
test_sentence_3 = ["You have a remarkable talent for solving complex problems with such ease."]
test_sentence_4 = ["The weather in London is often unpredictable and quite rainy during this season."]
test_sentence_5 = ["Please ensure that all the windows are closed before you leave the office."]

test1 = vectorizer.transform(test_sentence_1)
prediction = model.predict(test1)

print("칭찬" if prediction[0] == 1 else "일반")

test2 = vectorizer.transform(test_sentence_2)
prediction = model.predict(test2)

print("칭찬" if prediction[0] == 1 else "일반")

test3 = vectorizer.transform(test_sentence_3)
prediction = model.predict(test3)

print("칭찬" if prediction[0] == 1 else "일반")

test4 = vectorizer.transform(test_sentence_4)
prediction = model.predict(test4)

print("칭찬" if prediction[0] == 1 else "일반")

test5 = vectorizer.transform(test_sentence_5)
prediction = model.predict(test5)

print("칭찬" if prediction[0] == 1 else "일반")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. 학습 데이터 준비
# '칭찬' 키워드와 '일반 상황' 키워드를 학습 데이터에 고루 분포시킴
feature = [
    "The presentation was outstanding and professional",
    "You have a remarkable talent for solving problems",
    "Your work is excellent and great",
    "The weather is unpredictable and rainy",
    "Please ensure the windows are closed",
    "The store is often busy during this season"
]
labels = [1, 1, 1, 0, 0, 0]  # 1: 칭찬, 0: 일반

# 2. CountVectorizer를 활용한 단어 수 세기 (BoW)
cv = CountVectorizer()
cv_result = cv.fit_transform(feature)

# 단어 목록 및 인덱스 확인 (학습된 단어 사전)
print("--- [단어 사전 정보] ---")
print(f"전체 단어 목록: {cv.get_feature_names_out()}")
print(f"단어별 인덱스: {cv.vocabulary_}")
print("\n--- [문장별 벡터화 결과 (toarray)] ---")
print(cv_result.toarray())

# 3. 나이브 베이즈 모델 학습
model = MultinomialNB()
model.fit(cv_result, labels)

# 4. 테스트 데이터로 예측
# 학습 데이터에 포함된 단어들을 적절히 조합한 문장들
test_sentences = [
    "That was a professional presentation", # professional, presentation 포함
    "It is a rainy season",                 # rainy, season 포함
    "You have remarkable work"              # remarkable, work 포함
]

print("\n--- [예측 결과] ---")
for text in test_sentences:
    # 테스트 문장도 학습 시 사용한 Vectorizer로 변환해야 함
    test_vec = cv.transform([text])
    prediction = model.predict(test_vec)
    
    result = "칭찬" if prediction[0] == 1 else "일반"
    print(f"문장: '{text}' -> 결과: {result}")
```
